```python
# Create exports directory if it doesn't exist
dbutils.fs.mkdirs("/Volumes/pyspark_dbt/source/source_data/exports")

# Export daily revenue
spark.table("gold.daily_revenue") \
    .toPandas() \
    .to_csv("/Volumes/pyspark_dbt/source/source_data/exports/daily_revenue.csv", index=False)

# Export payment method KPI
spark.table("gold.payment_method_kpi") \
    .toPandas() \
    .to_csv("/Volumes/pyspark_dbt/source/source_data/exports/payment_method_kpi.csv", index=False)

# Export customer data
spark.table("gold.customer_360") \
    .toPandas() \
    .to_csv("/Volumes/pyspark_dbt/source/source_data/exports/customer_360.csv", index=False)

# Export failure analysis
spark.table("gold.payment_failures") \
    .toPandas() \
    .to_csv("/Volumes/pyspark_dbt/source/source_data/exports/payment_failures.csv", index=False)
```
